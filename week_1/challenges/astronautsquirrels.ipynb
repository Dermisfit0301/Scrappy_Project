# Challenge 3: Find the astronaut Scrappy Squirrels

In this challenge, we are going to make an automated parsing program where we can find the astronaut Scrappy Squirrels.
This program will go to the link and parse the data and extract it in form of excel sheet and then find the scrappy squirrels which are wearing astronaut equipment. Then the subset containing astronauts will be extracted in excel sheet.

```python
#getting packages
import requests
from openpyxl import Workbook
import pandas as pd
import os
import re
```

```python
# Function to fetch data from URL and return JSON content
def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch data from {url}. Status code: {response.status_code}")
        return None
```

```python
# Main function 
def main():
    base_url = "https://scrappyart.s3.ap-south-1.amazonaws.com/json/1"
    all_squirrels = []
 

    for page in range(1, 2001):
        url = f"{base_url}{page}"
        text_data = fetch_data(url)
        if text_data:
            all_squirrels.append(text_data)
            print(f"Fetched data from page {page}")
        else:
            print(f"Failed to fetch data from page {page}")
    
    return all_squirrels
```

```python
# Call the main function to fetch data and parse into a Pandas DataFrame
if __name__ == "__main__":
    data = main()
    for item in data:
        print(item)
```

```python
# Function to parse JSON entries and assign index numbers
def parse_json_entries(json_data):
    entries = re.split(r'(?<=})\n(?={)', json_data)  # Split entries using positive lookbehind and lookahead
    data = []
    for i, entry in enumerate(entries, 1):
        data.append({'index': i, 'entry': entry})
    return data
```

```python
# Assuming 'data' is a list of JSON strings
json_data = '\n'.join(data)
json_data
```

```python
# Parse JSON entries
parsed_data = parse_json_entries(json_data)
print(parsed_data)
```

```python
# converting to dataframe
df = pd.DataFrame(parsed_data)


print(df)
```

```python
#viewing new dataframe
df.head(5)
```

```python
# Splitting the 'entry' column into separate columns based on the patterns
split_data = df['entry'].str.extract(r'("name": "(.*?)")(?:.*?"description": "(.*?)")(?:.*"image": "(.*?)")(?:.*"attributes": (\[.*\]))?', expand=True)
split_data
```

```python
# Naming the columns
split_data.columns = ['index','name', 'description', 'image', 'attributes']
split_data
```

```python
#keeping only the necessary columns
new_df = split_data[['name', 'attributes']]
new_df
```

```python
# Filtering astronaut squirrels
ASTRONAUT_df = new_df[new_df['attributes'].str.contains('Astronaut|space', case=False)]
ASTRONAUT_df
```

```python
#number of astronaut squirrels
print('The number of squirrels who are astronauts are',len(ASTRONAUT_df))
```
