# XKCD compiler

In this module an automated parser has been created that will visit the link of XKCD comics and download the comics. then the comics will be compiled and saved in the form of a pdf. 


```python
import requests
import os
from PIL import Image
from fpdf import FPDF
```

```python
# Function to fetch information about all xkcd comics
def fetch_xkcd_comics_info():
    url = "https://xkcd.com/info.0.json"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        num_comics = min(data['num'], 2000)  
        return num_comics
    else:
        print("Failed to fetch xkcd comic information.")
        return 0
```

```python
# Function to download xkcd comic images
def download_xkcd_comics(num_comics):
    comic_dir = "xkcd_comics"
    os.makedirs(comic_dir, exist_ok=True)
    for i in range(1, num_comics + 1):
        url = f"https://xkcd.com/{i}/info.0.json"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            img_url = data['img']
            img_response = requests.get(img_url)
            if img_response.status_code == 200:
                with open(f"{comic_dir}/comic_{i}.png", 'wb') as f:
                    f.write(img_response.content)
                print(f"Comic {i} downloaded.")
            else:
                print(f"Failed to download comic {i}")
        else:
            print(f"Failed to fetch comic information for comic {i}")
```

```python
#main function
def main():
    num_comics = fetch_xkcd_comics_info()
    if num_comics > 0:
        download_xkcd_comics(num_comics)
        convert_to_pdf("xkcd_comics", "xkcd_comics.pdf")
        print("xkcd comic book created successfully.")
    else:
        print("Failed to fetch xkcd comic information.")
```

```python
#running the function
if __name__ == "__main__":
    main()
```

```python
# Function to convert images to PDF
def convert_to_pdf(comic_dir, output_pdf):
    pdf = FPDF()
    for filename in sorted(os.listdir(comic_dir)):
        if filename.endswith(".png"):
            try:
                pdf.add_page()
                pdf.image(os.path.join(comic_dir, filename), x=10, y=10, w=190)
                print(f"Added {filename} to PDF.")
            except RuntimeError as e:
                print(f"Error adding {filename} to PDF:", e)
            except Exception as e:
                print(f"Unexpected error adding {filename} to PDF:", e)
    pdf.output(output_pdf)
```

```python
# Provide the directory containing the PNG files and the output PDF filename
convert_to_pdf("xkcd_comics", "xkcd_comics.pdf")
```
